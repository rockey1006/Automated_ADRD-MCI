```python
import matplotlib.pyplot as plt
import numpy as np

# Define the results (metrics) for different models
labels = ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUROC', 'AUPRC']  # Names of the evaluation metrics
med_icd = [0.9305263157894736, 0.8584905660377359, 0.9285714285714286, 0.8921568627450982, 0.9682833499253359, 0.9517370723053938]  # MED-ICD- scores
med_icd_plus = [0.813953488372093, 0.7666666666666667, 0.9583333333333334, 0.8518518518518519, 0.875, 0.9033886436287362]  # MED-ICD+ scores
med_plus_icd = [0.8979591836734694, 0.7333333333333333, 0.9166666666666666, 0.8148148148148148, 0.9697297297297297, 0.9411778036778037]  # MED+ICD- scores
med_plus_icd_plus = [0.94106463878327, 0.7870370370370371, 0.9139784946236559, 0.8457711442786069, 0.973379026049815, 0.9264142214056826]  # MED+ICD+ scores

# Update the default font size for the plot
plt.rcParams.update({'font.size': 16})

x = np.arange(len(labels))  # Create an array for the x-axis positions based on the number of labels

# Create the figure for the bar chart with a specified size
plt.figure(figsize=(18, 10))

bar_width = 0.2  # Set the width of each bar
opacity = 0.8  # Set the opacity for the bars

# Plot the bars for each group, adjusting their position on the x-axis
plt.bar(x - 1.5 * bar_width, med_icd, bar_width, alpha=opacity, label='MED-ICD-')  # Plot bars for MED-ICD-
plt.bar(x - 0.5 * bar_width, med_icd_plus, bar_width, alpha=opacity, label='MED-ICD+')  # Plot bars for MED-ICD+
plt.bar(x + 0.5 * bar_width, med_plus_icd, bar_width, alpha=opacity, label='MED+ICD-')  # Plot bars for MED+ICD-
plt.bar(x + 1.5 * bar_width, med_plus_icd_plus, bar_width, alpha=opacity, label='MED+ICD+')  # Plot bars for MED+ICD+

# Set the labels for the axes
plt.xlabel('Metrics')  # X-axis label: metrics
plt.ylabel('Scores')  # Y-axis label: scores
plt.title('Comparison of Four Groups')  # Title of the plot

# Set the tick labels for the x-axis to show the metric names
plt.xticks(x, labels)

# Add a legend to the plot to identify each group
plt.legend()

# Annotate the bars with the actual values above them
for i, value in enumerate(med_icd):
    plt.text(i - 1.5 * bar_width, value + 0.02, f'{value:.2f}', ha='center')  # Annotate MED-ICD-
    plt.text(i - 0.5 * bar_width, med_icd_plus[i] + 0.02, f'{med_icd_plus[i]:.2f}', ha='center')  # Annotate MED-ICD+
    plt.text(i + 0.5 * bar_width, med_plus_icd[i] + 0.02, f'{med_plus_icd[i]:.2f}', ha='center')  # Annotate MED+ICD-
    plt.text(i + 1.5 * bar_width, med_plus_icd_plus[i] + 0.02, f'{med_plus_icd_plus[i]:.2f}', ha='center')  # Annotate MED+ICD+

# Set the y-axis limit to expand the range slightly above 1 for better visibility
plt.ylim(0, 1.2)

# Adjust layout to make sure labels and legends are properly displayed
plt.tight_layout()

# Save the plot as a high-resolution PNG file with tight bounding box

# Display the plot
plt.show()
```
